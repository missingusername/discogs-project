```python
from io import BytesIO
from bson import ObjectId
import pymongo
from PIL import Image
from IPython.display import display
from dotenv import load_dotenv
import os

def display_image_from_db(collection, document_id):
    # Retrieve the document by ID
    document = collection.find_one({"_id": document_id})

    if document and "image_data" in document:
        image_data = document["image_data"]
        image = Image.open(BytesIO(image_data))
        display(image)
    else:
        print("No image data found for the given document ID.")
```

```python
load_dotenv()

document_id = ObjectId("66f1dd68758ea48c44f7bf7f")
mongo_uri = os.getenv("MONGODB_URI")

client = pymongo.MongoClient(mongo_uri)
db = client["album_covers"]
sample_collection = db["5K-albums-sample"]
```

```python
display_image_from_db(sample_collection, document_id)
```
